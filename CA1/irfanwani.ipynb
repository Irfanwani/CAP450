```python
# Import pandas and numpy library.
# Upload the dataset ‘titanic.csv.’ Build a pandas dataframe using the provided dataset ‘titanic.csv’ to read the dataset. 
import pandas as pd
import numpy as np

data = pd.read_csv('titanic.csv')
df = pd.DataFrame(data)
df
```

```python
# Get the basic information about the dataset and display the basic statistical metrics such as count, mean, standard deviation, minimum, maximum, and percentiles.
basic = df.describe()
basic
```
