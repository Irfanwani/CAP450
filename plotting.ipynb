```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

file = pd.read_csv("bill.csv")

print(file.columns)

plt.bar(file["day"], file["reward"])
plt.title("Bar graph (day vs reward)")

plt.xlabel("Day")
plt.ylabel("Reward")
plt.show()

print(file["day"])
```

```python
# Bar graph using sns
sns.barplot(x='day', y='reward', data=file, hue='gender')
plt.title("Bar graph using SNS")

plt.show()
```

```python
sns.lineplot(data=file.drop(['total_bill'], axis=1))
plt.title("Line plot using sns")
plt.show()
```

```python
plt.plot(file['reward'])
plt.plot(file['person'])

plt.show()

file['person']
```

```python
data = [10, 23, 45, 12, 47]
courses = ["Maths", "AI", "DSA", "JAVA", "App dev"]
plt.figure(figsize=(5, 5))
plt.pie(data, labels=courses)

plt.show()
```

```python
cp = sns.color_palette('colorblind')
explode = [0.01, 0.2, 0.01, 0.01, 0.01]
plt.pie(data, labels=courses, colors=cp, autopct="%.2f%%", explode=explode)
plt.show()
```

```python
plt.hist(file['total_bill'])

plt.title("Histogram")
plt.show()

```

```python
sns.histplot(x='total_bill', data=file, kde=True, hue='gender')
plt.title("Histogram using Seaborn")
plt.show()
```

```python
plt.scatter(file['day'], file['reward'])
plt.xlabel("day")
plt.ylabel("reward")
plt.colorbar()
plt.title("Scatter graph using Matplotlib")
plt.show()
```

```python
sns.scatterplot(x='day', y='reward', data=file, hue='gender')
plt.title("Scatter plot using SNS(seaborn)")
plt.show()
```
